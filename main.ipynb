```python
# Import project modules
from ml.risk_predictor import RiskPredictor
from rules.student_copilot_rules import StudentCopilotRules
from planner.study_schedule_planner import Task, astar_schedule, schedule_to_text
```

```python
# Create a sample student profile and tasks
demo_profile = {
    'workload_hours': 12,
    'available_hours': 20,
    'stress_level': 5,
    'confidence_level': 7,
    'missed_sessions': 1,
    'sleep_hours': 7,
    'deadline_proximity': 'close'
}
demo_tasks = [
    Task("Math", 4, 3),
    Task("Biology", 3, 5),
    Task("History", 2, 4),
]
```

```python
# Run the ML predictor
ml = RiskPredictor()
ml.train_and_evaluate()
risk_level = ml.predict_student(demo_profile)
print("Predicted Risk Level:", risk_level)
```

```python
# Run the rule engine
rules = StudentCopilotRules()
inferred, trace = rules.evaluate_state(demo_profile)
print("Rule-based Explanation:", inferred)
print("Trace:")
for t in trace:
    print(t)
```

```python
# Run the planner
schedule = astar_schedule(demo_tasks)
print("Weekly Schedule:")
print(schedule_to_text(schedule))
```
